```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define function
def spitGaussian(mu, sigma, x):
    return((1/(sigma*np.sqrt(2*np.pi)))* np.exp(-((x-mu)**2)/(sigma**2)))
```

```python
# Define x as an array of numbers
x = np.linspace(-10,10,100)
```

```python
# Call Function:

Gauss1 = spitGaussian(3, 5, x)
Gauss2 = spitGaussian(10, -10, x)
#Gauss3 = (-5, 7, x)

print("Gaussians returned are")
print(Gauss1, Gauss2)
```

```python
# Plot
plt.plot(x, Gauss1, linestyle= '--', c ='g', label = 'Gauss1')
plt.plot(x, Gauss2, linestyle = '-', c='purple', label = 'Gauss2')
#plt.plot(x, Gauss3, marker = '>', c = 'purple')
plt.xlabel('x')
plt.ylabel('Gaussian')
plt.savefig('Gaussian.png')
plt.legend()
plt.show()
```
