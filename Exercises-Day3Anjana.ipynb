#### Exercise 1

There is a file in this directory called "histogram_exercise.dat" which consists of of randomly generated samples from a Gaussian distribution with an unknown $\mu$ and $\sigma$. Using what you've learned about fitting data, load up this file using np.genfromtxt, fit a Gaussian curve to the data and plot both the curve and the histogram of the data. As always, label everything, play with the colors, and choose a judicious bin size. 

Hint: if you attempt to call a function from a library or package that hasn't been imported, you will get an error.

```python
# your solution here

# your solution here
import numpy as np
import matplotlib.pyplot as plt
import scipy.optimize as opt # import your function here
from scipy.stats import norm

# load data file
Gauss_data = np.genfromtxt('../histogram_exercise.dat')
print(Gauss_data)

# Fit gaussian
mu, sigma = norm.fit(Gauss_data)
print(mu, sigma)

# Plot the fit and histogram
x = np.linspace(-5, 12, 1000)
pdf = norm.pdf(x, mu, sigma)

plt.plot(x, pdf, 'k-', linewidth=2)
plt.hist(Gauss_data, bins=50, color='darkblue', alpha=0.8, normed=True, fill=True, histtype='step', hatch = '//')
plt.title("Fit results: $\mu$ = %.2f, $\sigma$ = %.2f" % (mu, sigma)) 
#Pro-tip: this prints out the values of mu and sigma to 2
#decimal places.
plt.grid(True)
plt.show()
```

#### Exercise 2

Create a 1D interpolation along these arrays. Plot both the data (as points) and the interpolation (as a dotted line). Also plot the value of the interpolated function at x=325. What does the function look like to you?

```python
x = np.array([0., 50., 100., 150., 200., 250., 300., 350., 400., 450., 500])
y = np.array([0., 7.071, 10., 12.247, 14.142, 15.811, 17.321, 18.708, 20., 21.213, 22.361])

# solution here
#Import
from scipy.interpolate import interp1d

# Plot the data 
plt.plot(x, y, 'oy')
plt.show()

# Interpolate the data
interp = interp1d(x, y) # Calls function for x and y arrays; does not return anything
xnew = np.linspace(0, 500, 1000) #must completely fit inside the original array; Fit set completely within original data set
ynew = interp(xnew)

plt.plot(xnew, ynew, 'k--')
plt.plot(x, y, 'or')
plt.show()

#Value of interpolated function at x=325
yVal = interp(325)
print(yVal) # This is the value of y at x=325

#Plot the value-  point is in blue
plt.plot(xnew, ynew, 'k--')
plt.plot(325, 18.01, marker ='h', markersize = 10, c='b') #The value
plt.plot(x, y, 'or')
plt.show()
# This function looks like a logarithmic function
```
